```python
import json
import os
from PIL import Image
import pandas as pd
```

```python
DATABASE_FOLDER = '../data/wider_attribute_image'
JSON_FILE = '../data/wider_attribute_annotation/wider_attribute_trainval.json'
DATA_FOLDER = '../data/wider_attribute_image/Image'
CROPPED_FOLDER = '../data/wider_attribute_image/Cropped'
```

```python
os.mkdir(os.path.join(CROPPED_FOLDER))
os.mkdir(os.path.join(CROPPED_FOLDER, 'train'))
os.mkdir(os.path.join(CROPPED_FOLDER, 'train/-1'))
os.mkdir(os.path.join(CROPPED_FOLDER, 'train/0'))
os.mkdir(os.path.join(CROPPED_FOLDER, 'train/1'))
os.mkdir(os.path.join(CROPPED_FOLDER, 'val'))
os.mkdir(os.path.join(CROPPED_FOLDER, 'val/-1'))
os.mkdir(os.path.join(CROPPED_FOLDER, 'val/0'))
os.mkdir(os.path.join(CROPPED_FOLDER, 'val/1'))
```

```python
with open(JSON_FILE) as f:
    data = json.load(f)
```

```python
marking = pd.DataFrame(columns = ['filename', 'selection', 'width', 'height', 'sex'])
marking
```

```python
# формирование базы данных
numb = 0
for elem in data['images']: # каждый элемент - словарь
    img_name = elem['file_name']
    img = Image.open(os.path.join(DATA_FOLDER, img_name))
    for annotation in elem['targets']: # каждый элемент - словарь
        sex = annotation['attribute'][0]
        bbox = tuple(map(int, annotation['bbox']))
        frame = (bbox[0], bbox[1], bbox[0]+bbox[2], bbox[1]+bbox[3])
        cropped_img = img.crop(frame)
        if 'train/' in img_name:
            cropped_path = os.path.join(os.path.join(CROPPED_FOLDER, 'train'), str(sex))
        else:
            cropped_path = os.path.join(os.path.join(CROPPED_FOLDER, 'val'), str(sex))
        cropped_name = os.path.splitext(os.path.basename(img_name))[0] + '_' + str(numb) + '.png'
        cropped_img.save(os.path.join(cropped_path, cropped_name))
        width, height = cropped_img.size
        numb += 1
```

```python
# статистика по размерам картинок
for root, d, files in os.walk(CROPPED_FOLDER):
    for file in files:
        img = Image.open(os.path.join(os.path.join(CROPPED_FOLDER, root), file))
        if 'train\\' in root:
            selection = 'train'
        else:
            selection = 'val'
        if '\\-1' in root:
            sex = -1
        elif '\\0' in root:
            sex = 0
        else:
            sex = 1
        width, height = img.size
        marking.loc[len(marking)] = [file, selection, width, height, sex]
```

```python
marking.to_csv(os.path.join(DATABASE_FOLDER, 'cropped_marking.csv'), index = False)
```
