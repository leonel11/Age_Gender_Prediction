```python
import pandas as pd
import os
from PIL import Image
```

```python
DATABASE_FOLDER = 'C:/Users/User/Downloads/wider_attribute_image'
IMAGES_FOLDER = os.path.join(DATABASE_FOLDER, 'Cropped')
MARKING_FILE = os.path.join(DATABASE_FOLDER, 'cropped_marking.csv')
DATABASE_COLUMNS = ['filename', 'selection', 'width', 'height', 'sex']
RESIZED_DATABASE_COLUMNS = ['filename', 'selection', 'sex']
RESIZED_FOLDER = os.path.join(DATABASE_FOLDER, 'Resized')
RESIZED_MARKING_FILE = os.path.join(DATABASE_FOLDER, 'resized_marking.csv')
```

```python
os.mkdir(os.path.join(RESIZED_FOLDER))
os.mkdir(os.path.join(RESIZED_FOLDER, 'train'))
os.mkdir(os.path.join(RESIZED_FOLDER, 'train/-1'))
os.mkdir(os.path.join(RESIZED_FOLDER, 'train/0'))
os.mkdir(os.path.join(RESIZED_FOLDER, 'train/1'))
os.mkdir(os.path.join(RESIZED_FOLDER, 'val'))
os.mkdir(os.path.join(RESIZED_FOLDER, 'val/-1'))
os.mkdir(os.path.join(RESIZED_FOLDER, 'val/0'))
os.mkdir(os.path.join(RESIZED_FOLDER, 'val/1'))
```

```python
marking = pd.read_csv(MARKING_FILE)
marking.head()
```

```python
len(marking)
```

```python
i = 0
while (i < len(marking)):
    row = marking.iloc[i]
    if ((row['width'] > 1.5*row['height']) or (row['width']) < 25):
        print(row['filename'])
        os.remove(os.path.join(IMAGES_FOLDER, os.path.join(row['selection'], os.path.join(str(row['sex']), row['filename']))))
        marking = marking.drop(marking.index[i])
    else:
        i += 1
```

```python
len(marking)
```

```python
i = 0
while (i < len(marking)):
    row = marking.iloc[i]
    coef = float(row['height']/row['width'])
    new_width, new_height = 70, int(70*coef)
    filename = os.path.join(row['selection'], os.path.join(str(row['sex']), row['filename']))
    if (new_height < 60):
        os.remove(os.path.join(IMAGES_FOLDER, filename))
        marking = marking.drop(marking.index[i])
    else:
        row['width'], row['height'] = new_width, new_height
        img = Image.open(os.path.join(IMAGES_FOLDER, filename))
        img = img.resize((new_width, new_height), Image.ANTIALIAS)
        img.save(os.path.join(RESIZED_FOLDER, filename))
        i += 1
```

```python
marking[RESIZED_DATABASE_COLUMNS].to_csv(RESIZED_MARKING_FILE, index = False)
```
