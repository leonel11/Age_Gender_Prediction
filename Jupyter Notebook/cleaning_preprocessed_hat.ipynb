```python
import pandas as pd
import os
from PIL import Image
```

```python
DATABASE_FOLDER = 'C:/Users/User/Downloads/hatdb'
IMAGES_FOLDER = os.path.join(DATABASE_FOLDER, 'cropped')
MARKING_FILE = os.path.join(DATABASE_FOLDER, 'cropped_marking.csv')
DATABASE_COLUMNS = ['filename', 'width', 'height', 'sex']
RESIZED_DATABASE_COLUMNS = ['filename', 'sex', 'elderly', 'middleaged', 'young', 'teen', 'kid', 'baby']
RESIZED_FOLDER = os.path.join(DATABASE_FOLDER, 'resized')
RESIZED_MARKING_FILE = os.path.join(DATABASE_FOLDER, 'resized_marking.csv')
```

```python
os.mkdir(RESIZED_FOLDER)
os.mkdir(os.path.join(RESIZED_FOLDER, 'F'))
os.mkdir(os.path.join(RESIZED_FOLDER, 'M'))
```

```python
marking = pd.read_csv(MARKING_FILE)
marking.head()
```

```python
len(marking)
```

```python
i = 0
while (i < len(marking)):
    row = marking.iloc[i]
    if ((row['width'] > 1.5*row['height']) or (row['width']) < 25):
        print(row['filename'])
        os.remove(os.path.join(IMAGES_FOLDER, row['filename']))
        marking = marking.drop(marking.index[i])
    else:
        i += 1
```

```python
len(marking)
```

```python
i = 0
while (i < len(marking)):
    row = marking.iloc[i]
    coef = float(row['height']/row['width'])
    new_width, new_height = 70, int(70*coef)
    filename = row['filename']
    if (new_height < 60):
        os.remove(os.path.join(IMAGES_FOLDER, filename))
        marking = marking.drop(marking.index[i])
    else:
        row['width'], row['height'] = new_width, new_height
        img = Image.open(os.path.join(IMAGES_FOLDER, filename))
        img = img.resize((new_width, new_height), Image.ANTIALIAS)
        if row['sex'] == 1:
            sub_folder = 'F'
        else:
            sub_folder = 'M'
        img.save(os.path.join(RESIZED_FOLDER, os.path.join(sub_folder, filename)))
        i += 1
```

```python
marking[RESIZED_DATABASE_COLUMNS].to_csv(RESIZED_MARKING_FILE, index = False)
```
